```python
import cv2
import numpy as np
import sklearn 
```

```python
recognition =cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
cap= cv2.VideoCapture(0)
data=[]
while True:
    k , video = cap.read()
    colour = cv2.cvtColor(video,cv2.COLOR_BGR2GRAY)
    person =recognition.detectMultiScale(video)
    for x,y,w,h in person :
        cv2.rectangle(video,(x,y),(x+w,y+h),(0,0,255),8)
        in_size=video[y:y+h,x:x+w,:]
        in_size=cv2.resize(in_size,(50,50))
        print(len(data))
        if len(data)<300 :
            data.append(in_size)
    cv2.imshow("Image",video)
    if cv2.waitKey(20)==ord('q') or len(data)>=300:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
np.save('without mask',data)
```

```python
np.save('with mask',data)
```
